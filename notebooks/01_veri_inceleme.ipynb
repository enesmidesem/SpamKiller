```python
import pandas as pd

# Veriyi yükle
df = pd.read_csv('../data/SMSSpamCollection', 
                 sep='\t',          # sütunlar tab ile ayrılmış
                 header=None,       # başlık satırı yok
                 names=['etiket', 'metin'])  # sütun isimlerini biz veriyoruz

# İlk 5 satırı gör
df.head()
```

```python
# Kaç spam, kaç gerçek var?
print(df['etiket'].value_counts())

# Etiketleri sayıya çevir (spam=1, ham=0)
df['label'] = df['etiket'].map({'spam': 1, 'ham': 0})

print(f"\nToplam kayıt: {len(df)}")
print(f"Spam: {df['label'].sum()}")
print(f"Gerçek: {(df['label'] == 0).sum()}")
```

```python
from deep_translator import GoogleTranslator
import time

translator = GoogleTranslator(source='en', target='tr')

def cevir(metin):
    try:
        return translator.translate(str(metin)[:500])  # 500 karakter sınırı
    except:
        return metin  # hata olursa orijinali kullan

# Önce küçük bir test yapalım (ilk 5 satır)
df_test = df.head(5).copy()
df_test['metin_tr'] = df_test['metin'].apply(lambda x: cevir(x))
print(df_test[['metin', 'metin_tr']])
```

```python
# DİKKAT: Bu 10-15 dakika sürebilir!
df['metin_tr'] = df['metin'].apply(lambda x: cevir(x))

# Kaydet (bir daha çevirmek zorunda kalmayalım)
df.to_csv('../data/spam_turkce.csv', index=False)
print("Kaydedildi!")
```
