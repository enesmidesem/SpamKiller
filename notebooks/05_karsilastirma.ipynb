```python
import matplotlib.pyplot as plt
import numpy as np

modeller = ['Lojistik\nRegresyon', 'Random\nForest', 'BERTurk']

metrikler = {
    'Accuracy':  [96.77, 98.47, 98.56],
    'Precision': [92.48, 99.25, 93.46],
    'Recall':    [82.55, 89.26, 95.97],
    'F1 Score':  [87.23, 93.99, 94.70],
}

renkler = ['#4C72B0', '#55A868', '#C44E52']
x = np.arange(len(modeller))
genislik = 0.2

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('SpamKiller — Model Karşılaştırması', fontsize=16, fontweight='bold', y=1.02)

for ax, (metrik_adi, degerler) in zip(axes.flatten(), metrikler.items()):
    bars = ax.bar(x, degerler, color=renkler, width=0.5, edgecolor='white', linewidth=1.2)
    
    # Değerleri barların üstüne yaz
    for bar, deger in zip(bars, degerler):
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.3,
                f'%{deger}', ha='center', va='bottom', fontsize=11, fontweight='bold')
    
    ax.set_title(metrik_adi, fontsize=13, fontweight='bold')
    ax.set_xticks(x)
    ax.set_xticklabels(modeller, fontsize=11)
    ax.set_ylim(75, 102)
    ax.set_ylabel('Skor (%)')
    ax.grid(axis='y', alpha=0.3)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.savefig('../data/model_karsilastirma.png', dpi=150, bbox_inches='tight')
plt.show()
print("Grafik kaydedildi ✓")
```
